Question no. 1 What are the most common reasons for missing data in ETL pipelines?

Answer: Missing data in ETL pipelines can occur due to several reasons:

1 Human Error: Users may skip fields while entering data.

2. system Failures: Network issues, server crashes, or application  errors can prevent data collection.

3. Data integration Problem: Different systems may have different formats, causing some values to be lost during extraction or transformation.

4. Optional fields: Customers may choose not to provide information such as income or phone number.

5. Data corruption: Files may become damaged during transfer or storage.

6. Migration issues: Some records may be lost when moving data between systems.

Question no. 2 Why is blindly deleting rows with missing values considered a bad practice in ETL?

Answer: Blindly deleting rows can lead to:
1. Loss of valuable information.

2. Reduced dataset size, affecting analysis accuracy.

3. Biased results if missing values follow a pattern.

4. Loss of important customers or transactions.

Question no. 3. Explain the difference between:
Listwise deletion

column deletion

Also mention one scenario where each is appropriate.

Answer:  Listwise deletion  

1. Removes entire rows containing missing values

2. Loss of records(rows)

3. Reduces number of observations

4. When only a few rows are missing

Scenario for Listwise Deletion

A dataset of 10,000 customers has only 20 records with missing Region values. Deleting those 20 rows will have minimal impact.

Column Deletion

1. Removes an entire column containing many missing values

2. Loss of a feature(column)

3. Reduces number of variables

4. When a column has to many missing values

Scenario for Column Deletion

A column named "Middle_Name" has 95% missing values and is not important for analysis. The entire column can be removed.

Question no. 4 Why is median imputation preferred over mean imputation for skewed data such as income?

Answer: Income data is usually skewed because a few people earn extremely high incomes.

Mean is affected by extreme values (outliers).
Median is not affected by outliers and better represents the typical value.

Question no. 5 What is forward fill and in what type of dataset is it most useful?

Answer: Forward fill replaces a missing value with the most recent non-missing value that appears before it.

Forward fill is useful for time-series data such as stock prices, sensor reading, inventory levels, and sales reports.
